### 4)  For each quantitative attribute, calculate its average, standard deviation, minimum, mode, range and maximum values.



```python
from pandas.api.types import is_numeric_dtype

for col in data.columns:
    if is_numeric_dtype(data[col].dtype):
        print(f"{col} :")
        print("\tMax :", data[col].max())
        print("\tMin :", data[col].min())
        print("\tStandard Deviation :", data[col].std())
        print("\tAverage :", data[col].mean())
        print("\tRange :", data[col].max() - data[col].min())
        
        if col != "PassengerId":
            print("\tMode :",data[col].mode()[0])
```
